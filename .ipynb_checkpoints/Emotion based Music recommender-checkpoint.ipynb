```python
import pandas as pd
```

```python
data=pd.read_csv('dataset.csv')
data
```

```python
data.drop_duplicates(inplace=True,subset=['track_name'])
```

```python
name=data['track_name']
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
col_features = ['emotion', 'energy', 'valence', 'loudness']
X = MinMaxScaler().fit_transform(data[col_features])
kmeans = KMeans(init="k-means++",
                n_clusters=5,
                random_state=15).fit(X)
data['kmeans'] = kmeans.labels_
```

```python
data['song_name']=name
```

```python
data
```

```python
cluster=data.groupby(by=data['kmeans'])
```

```python
y=data.pop('kmeans')
x=data.drop(columns=['track_name','album','track_id','song_name'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
!pip install lightgbm
from lightgbm import LGBMClassifier
model=LGBMClassifier().fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
df=cluster.apply(lambda x: x.sort_values(["popularity"],ascending=False))
df.reset_index(level=0, inplace=True)
```

```python
from keras.preprocessing.image import img_to_array
import imutils
from keras.models import load_model
import numpy as np
import cv2
```

```python
detection_model_path = 'haarcascade_frontalface_default.xml'
emotion_model_path = 'final_model3.h5'
face_detection = cv2.CascadeClassifier(detection_model_path)
emotion_classifier = load_model(emotion_model_path, compile=False)
EMOTIONS = ["happy","sad"]
```

```python
from keras.preprocessing import image
def emotion_testing():
        cap=cv2.VideoCapture(0)
        while True:
            ret,test_img=cap.read()# captures frame and returns boolean value and captured image
            if not ret:
                continue
            gray_img= cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)

            faces_detected = face_detection.detectMultiScale(gray_img, 1.32, 5)


            for (x,y,w,h) in faces_detected:
                cv2.rectangle(test_img,(x,y),(x+w,y+h),(255,0,0),thickness=7)
                roi_gray=gray_img[y:y+w,x:x+h]#cropping region of interest i.e. face area from  image
                roi_gray=cv2.resize(roi_gray,(48,48))
                img_pixels = image.img_to_array(roi_gray)
                img_pixels = np.expand_dims(img_pixels, axis = 0)
                img_pixels /= 255

                predictions = emotion_classifier.predict(img_pixels)

                #find max indexed array
                max_index = np.argmax(predictions[0])
                predicted_emotion = EMOTIONS[max_index]

                cv2.putText(test_img, predicted_emotion, (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)

            resized_img = cv2.resize(test_img, (1000, 700))
            cv2.imshow('Facial emotion analysis ',resized_img)



            if cv2.waitKey(0) & 0xFF == ord('q'):
                     break
        cap.release()
        cv2.destroyAllWindows
        return predicted_emotion
```

```python
emotion_word=emotion_testing()
if emotion_word=='Sad':
    emotion_code=0
elif emotion_word=='Happy':
    emotion_code=1
elif emotion_word=='Angry':
    emotion_code=2
elif emotion_word=='Fear':
    emotion_code=3
else
    emotion_code=4
```

```python
def get_results(emotion_code):
    NUM_RECOMMEND=10
    happy_set=[]
    sad_set=[]
    Angry_set=[]
    Fear_set[=[]
    Neutral_set[]
    if emotion_code==0:
        sad_set.append(df[df['kmeans']==0]['song_name'].head(NUM_RECOMMEND))
        return pd.DataFrame(sad_set).T
    elif emotion_code==1:
        happy_set.append(df[df['kmeans']==1]['song_name'].head(NUM_RECOMMEND))
        return pd.DataFrame(happy_set).T
    elif emotion_code==2:
        Angry_set.append(df[df['kmeans']==2]['song_name'].head(NUM_RECOMMEND))
        return pd.DataFrame(Angry_set).T
    elif emotion_code==3:
        Fear_set.append(df[df['kmeans']==3]['song_name'].head(NUM_RECOMMEND))
        return pd.DataFrame(Fear_set).T
    else:
        Neutral_set.append(df[df['kmeans']==4]['song_name'].head(NUM_RECOMMEND))
        return pd.DataFrame(Neutral_set).T
```
